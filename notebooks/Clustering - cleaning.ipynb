```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
gdf_delitos = gpd.read_file("../output/tweets-delitos-localizados.geojson")
```

```python
df = gdf_delitos.drop(columns=['geometry', 'Unnamed: 0', "index_right", "stroke", "stroke-width", "stroke-opacity", "fill", "fill-opacity", "Description"])  # Elimina la columna de geometría
```

```python
df["Date"] = df["Timestamp"].dt.date
```

```python
df["Time"] = df["Timestamp"].dt.time
```

```python
df.drop(columns="Timestamp", inplace=True)
```

```python
df
```

```python
df.drop_duplicates(subset=['Content'], inplace=True)
```

```python
df.reset_index(drop= True, inplace = True)
```

```python
keyword_dict = {
    'Robo': ['robo', 'hurto', 'atraco', 'saqueo', 'apropiacion indebida','asalt','roba'],
    'Agresión': ['agresion', 'pelea', 'violencia', 'golpiza', 'ataque', 'maltrato','balacera','hirio','heri','golp','bala','puñetes', 'dispar', 'paliza', 'balead','gresion' ],
    'Drogas': ['drogas', 'narcotrafico', 'estupefacientes', 'consumo', 'venta'],
    'Homicidio': ['homicidio', 'muert', 'asesin', 'sicariato','masacre','desmembr', 'decapit', 'acribill', 'descuarti', 'femicidio', 'mata', 'mato', ],
    'Accidente Vehicular': ['atropella','accidente', 'colision', 'choque', 'atropello', 'vehiculo', 'conductor','siniestro de transito','volco','volcamiento','volcado'],
    'Secuestro': ['secuestr', 'retenido', 'plagio', 'liberacion', 'rescate', 'encerr'],
    'Fraude': ['fraude', 'estafa', 'engaño', 'falsificacion', 'delito financiero', 'extorsionadores'],
    'Extorsión':['vacuna', 'extorsion'],
    'Incendio': ['incendio', 'fuego', 'quemar', 'combustion', 'siniestro'],
    'Vandalismo': ['vandalismo', 'destrozo', 'daños', 'destruccion', 'pintadas','explosivo','bomba','relaciones sexuales', 'dinamita'],
    'Violación': ['violacion', 'agresion sexual', 'abuso sexual', 'abusar', 'violador'],
    # Agrega más tipos de delitos y palabras clave según sea necesario
}
```

```python
def remove_tildes(texto):
    tildes_dict = {
        'á': 'a',
        'é': 'e',
        'í': 'i',
        'ó': 'o',
        'ú': 'u',
        'Á': 'A',
        'É': 'E',
        'Í': 'I',
        'Ó': 'O',
        'Ú': 'U',
    }

    texto_sin_tildes = ''.join(tildes_dict.get(caracter, caracter) for caracter in texto)
    return texto_sin_tildes

def categorizar_delito(tweet):
    contenido = remove_tildes(tweet['Content'].lower())
    
    # Buscar palabras clave en el contenido del tweet
    for delito, palabras_clave in keyword_dict.items():
        for palabra in palabras_clave:
            if palabra in contenido:
                return delito
    
    # Si no se encuentra ninguna coincidencia, asignar 'Otro'
    return 'Otro'
```

```python
df['TipoDelito'] = df.apply(categorizar_delito, axis=1)
```

```python
df.loc[df["TipoDelito"]=="Otro",:].shape
```

```python
def cambiar_Tipo_Delito(indices, tipo_delito):
    for indice in indices:
        df.at[(indice-2),'TipoDelito'] = tipo_delito
```

```python
cambiar_Tipo_Delito([47,144,516], 'Robo')
cambiar_Tipo_Delito([305,352,462], 'Homicidio')
cambiar_Tipo_Delito([189,265,563,564,565], 'Vandalismo')
cambiar_Tipo_Delito([526], 'Extorsión')
```

```python
df = df.loc[df["TipoDelito"]!="Otro",:]
```

```python
df_interactions = df.dropna(subset="DisplayName")
```

```python
df.to_csv("../output/tweets_con_tipo_delito.csv", index=False)
```

```python
interXdelito = df_interactions.groupby("TipoDelito")["Comments", "Likes", "Retweets", "Interactions"].mean().reset_index()
```
